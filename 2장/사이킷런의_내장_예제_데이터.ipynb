```python
from sklearn.datasets import load_iris
```

```python
iris_data = load_iris()
print(type(iris_data))
```

```python
keys = iris_data.keys()
print('붓꽃 데이터 세트의 키들:', keys)
```

- `data`는 피처의 데이터셋을 나타냄
- `target`은 분류 시 레이블 값, 회귀일 때는 숫자 결괏값 데이터셋을 나타냄
- `target_names`는 개별 레이블의 이름을 나타냄
- `feature_names`는 피처의 이름을 나타냄
- `DESCR`은 데이터셋에 대한 설명과 각 피처의 설명을 나타냄

```python
print('feature_names 의 type:', type(iris_data.feature_names))
print('feature_names 의 shape:', len(iris_data.feature_names))
print(iris_data.feature_names)

print('\ntarget_names 의 type:', type(iris_data.target_names))
print('target_names 의 shape:', len(iris_data.target_names))
print(iris_data.target_names)

print('\ndata 의 type:', type(iris_data))
print('data 의 shape:', iris_data.data.shape)
print(iris_data['data'])

print('\ntarget 의 type:', type(iris_data.target))
print('target 의 shape:', iris_data.target.shape)
print(iris_data.target)
```
